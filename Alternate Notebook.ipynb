```python
import pandas as pd
```

```python
# WORKING ON FIRST FILE FOR FOOD RESTAURANTS
#
#
```

```python
# load the data for food restaurants
fast_food_df = pd.read_csv("Resources/Datafiniti_Fast_Food_Restaurants_May19.csv")
fast_food_df.head()
```

```python
# keep the columns we need
fast_food_df_reduced = fast_food_df[["name","city","province","postalCode"]]
fast_food_df_reduced.head()
```

```python
# rename the columns to something more appropriate
fast_food_df_renamed = fast_food_df_reduced.rename(columns={
    "name":"restaurant_name",
    "province":"state",
    "postalCode":"old_zipcode"
})
fast_food_df_renamed.head()
```

```python
#sort the values by zipcode, and reset index
fast_food_sort_df = fast_food_df_renamed.sort_values("old_zipcode").reset_index()
# exclude old index
fast_food_sort_index = fast_food_sort_df[["restaurant_name","city","state","old_zipcode"]]
fast_food_sort_index
```

```python
# splitting the zip code with the first 5 digits and the additional 4 identifiers
fast_food_split = fast_food_sort_index["old_zipcode"].str.split("-", expand = True)
fast_food_split.head()
```

```python
# rename the dataframe
fast_food_zip_rename = fast_food_split.rename(columns ={0 : "zipcode", 1: "extra 4 digits"})
fast_food_zip_rename.head()
```

```python
# drop the extra 4 digits
fast_food_newzipcodes = fast_food_zip_rename['zipcode'].to_frame(name = 'zipcode')
food_zipcodes = fast_food_newzipcodes[["zipcode"]]
food_zipcodes
```

```python
# change the data type from object to int64 for future merging
food_zipcodes= food_zipcodes["zipcode"].astype(str).astype(int)
```

```python
# join the dataframe of the new split zipcodes with the original dataframe
join_food_frame = fast_food_sort_index.join(food_zipcodes, how='outer')
join_food_frame
```

```python
# only keep the columns that we need for merging with the other csv dataframes, we will merge by zipcode
final_food_df = join_food_frame[["restaurant_name" , "city" , "state" , "zipcode"]]
final_food_df
```

```python
# WORKING ON SECOND FILE FOR US POPULATION
#
#
```

```python
# load the census data
population_df = pd.read_csv("Resources/population_by_zip_2010.csv")
population_df.head()
```

```python
# sort the population by zipcode
population_sort_df = population_df.sort_values("zipcode")
```

```python
# keep only the columns we need
pop_reduced = population_sort_df[["zipcode","population"]]
pop_reduced.head()
```

```python
# group by zipcode
pop_grouped = pop_reduced.groupby(["zipcode"])
pop_grouped.head()
```

```python
# sum the population by the zipcode, and create a dataframe with zipcode and population
final_population_df = pop_grouped['population'].sum().to_frame(name = 'population').reset_index()
final_population_df
```

```python
#
#
#
```

```python
# TIME TO MERGE BOTH FINAL DATA SETS FROM THE TWO CSV FILES INTO ONE BY ZIPCODE
#
#
```

```python
# Merge the two DataFrames together based on the zipcodes they share
two_merged_df = pd.merge(final_food_df,final_population_df, on="zipcode")
two_merged_df
```

```python
# let us drop any columns with missing information
two_merged_df_drop = two_merged_df.dropna(how = 'any')

# at this point the transformation process is complete. Now the dataframe can be
# used to perform the desired analysis
two_merged_df_drop
```
